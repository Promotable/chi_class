# Joining/Aggregating

```python
# join
neighborhood_area = pd.DataFrame({
    'neighborhood': ['South Addison', 'Girdwood', 'Sand Lake', 'Spenard', 'Turnagain', 'Fairview'],
    'area': [1, 2, 2, 1, 1, 2]})

airbnb_area = pd.merge(airbnb, neighborhood_area, on = 'neighborhood', how = 'left')
```

```python
# sorting
airbnb.sort_values(['host_id', 'room_id'], ascending=False)
```

```python
# groupby
airbnb.groupby(['neighborhood', 'room_type'])['bedrooms'].mean()
```
